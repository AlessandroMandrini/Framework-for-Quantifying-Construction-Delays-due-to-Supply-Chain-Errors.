```python
import numpy as np
# Define buffer variables for each component using interactive parameters
Buffer_IGUs = 0.3
Buffer_AluminumProfiles = 0.4
Buffer_Fasteners = 0.5
Buffer_Gaskets = 0.5
Buffer_ChemicalAnchors = 0.2
Buffer_ShadingSystems = 0.3
Buffer_SpandrelPanels = 0.5
Buffer_Brackets = 0.5
Buffer_Assembly = 0.5

# Define additional parameters
recovery_installation = 0.5 # @param {type:"number"}
```

```python
import pandas as pd
import numpy as np
from scipy.stats import beta
import os

# Define the number of simulations
num_simulations = 5000

# List of CSV files to process
csv_files = [
    'ErrorBreakdown_IGUs.csv',
    'ErrorBreakdown_AluminumProfiles.csv',
    'ErrorBreakdown_Fasteners.csv',
    'ErrorBreakdown_Gaskets.csv',
    'ErrorBreakdown_ChemicalAnchors.csv',
    'ErrorBreakdown_ShadingSystems.csv',
    'ErrorBreakdown_SpandrelPanels.csv',
    'ErrorBreakdown_Brackets.csv',
    'ErrorBreakdown_Assembly.csv',
    'ErrorBreakdown_Installation.csv' # Added Installation CSV
]

# Let's re-define the generation functions here for clarity and self-containment.
def generate_delay(row):
    if row['event_occurred']:
        mean = row['delay_av']
        min_val = row['delay_min']
        max_val = row['delay_max']

        if min_val == max_val:
            return round(float(min_val), 1) # Return as float with one decimal

        kappa = 13
        normalized_mean = (mean - min_val) / (max_val - min_val)
        epsilon = 1e-6
        alpha = normalized_mean * kappa + epsilon
        beta_param = (1 - normalized_mean) * kappa + epsilon

        beta_val = np.random.beta(alpha, beta_param)
        generated_val = min_val + beta_val * (max_val - min_val)
        return round(np.clip(generated_val, min_val, max_val), 1) # Round to one decimal
    else:
        return 0.0 # Return 0.0 as float

print(f"Starting {num_simulations} full simulations.")

# Store component names for later printing
component_names_list = []
# Preload templates to avoid re-reading CSVs in each simulation run
component_templates = {}
for csv_file in csv_files:
    component_name = os.path.basename(csv_file).replace('ErrorBreakdown_', '').replace('.csv', '')
    component_names_list.append(component_name)
    original_df_template = pd.read_csv(csv_file, delimiter=';')
    # Use 'Error' column as ID, or 'Error / Disruption' if 'Error' is not found
    if 'Error' in original_df_template.columns:
        original_df_template['ID'] = original_df_template['Error']
    elif 'Error / Disruption' in original_df_template.columns:
        original_df_template['ID'] = original_df_template['Error / Disruption']
    else:
        raise KeyError(f"Neither 'Error' nor 'Error / Disruption' column found in {csv_file}")

    # Check and add 'Minimum', 'Average', 'Maximum' columns if they don't exist, filling with 0
    for col in ['Minimum', 'Average', 'Maximum']:
        if col not in original_df_template.columns:
            original_df_template[col] = 0.0 # Fill with 0.0 for consistent type
        # Convert the column to numeric, handling potential string format with comma
        original_df_template[col] = original_df_template[col].astype(str).str.replace(',', '.').astype(float)

    original_df_template['P_err'] = original_df_template['Probability of error'].str.replace(',', '.').str.strip('%').astype(float) / 100
    original_df_template = original_df_template.rename(columns={
        'Minimum': 'delay_min',
        'Average': 'delay_av',
        'Maximum': 'delay_max'
    })
    component_templates[component_name] = original_df_template

all_simulations_results = [] # List to store results of each full simulation

# Lists to store value_to_subtract for plotting
all_value_to_subtract = []
all_value_to_subtract_assembly = []

for s_idx in range(num_simulations):
    current_simulation_delays = {}
    current_simulation_corrected_delays = {}

    # Process each component for the current simulation run
    for component_name in component_names_list:
        original_df_template = component_templates[component_name]

        # Create a fresh copy of the original data for this simulation run
        current_df = original_df_template.copy()

        # Step 1: Generate event_occurred
        current_df['event_occurred'] = np.random.rand(len(current_df)) <= current_df['P_err']

        # Step 2: Generate delay
        current_df['delay'] = current_df.apply(generate_delay, axis=1)

        # Step 3: Calculate total_delay for this component
        total_delay = current_df['delay'].sum()

        current_simulation_delays[component_name] = total_delay

        # Apply buffer logic
        original_delay = total_delay
        buffer_var_name = f'Buffer_{component_name}'
        buffer_value = globals().get(buffer_var_name, 0.0) # Ensure buffer_value is retrieved

        if component_name == 'Assembly':
            # For Assembly, corrected delay needs Delay_PreAssembly, which is calculated later.
            # We'll set it as a placeholder and correct it later.
            current_simulation_corrected_delays[component_name] = original_delay
        elif component_name == 'Installation':
            # User request: Delay_Installation should be calculated but not corrected by buffer.
            current_simulation_corrected_delays[component_name] = original_delay
        else:
            if original_delay <= buffer_value:
                current_simulation_corrected_delays[component_name] = 0.0 # Changed to 0.0 as requested
            else: # original_delay > buffer_value
                if buffer_value <= 0:
                    random_beta_val = np.random.beta(a=1, b=3)
                    value_to_subtract = buffer_value + random_beta_val * buffer_value
                    corrected_delay = max(0.0, original_delay - value_to_subtract)
                    current_simulation_corrected_delays[component_name] = round(corrected_delay, 1)
                    all_value_to_subtract.append(value_to_subtract) # Collect for plotting
                else:
                    random_beta_val = np.random.beta(a=1, b=3)
                    value_to_subtract = buffer_value + random_beta_val * buffer_value
                    corrected_delay = max(0.0, original_delay - value_to_subtract)
                    current_simulation_corrected_delays[component_name] = round(corrected_delay, 1)
                    all_value_to_subtract.append(value_to_subtract) # Collect for plotting

    # Now calculate composite delays for the current simulation
    components_for_preassembly = [
        'IGUs',
        'AluminumProfiles',
        'Fasteners',
        'Gaskets',
        'ShadingSystems',
        'SpandrelPanels'
    ]

    corrected_delays_to_consider = []
    for comp_name in components_for_preassembly:
        corrected_delays_to_consider.append(current_simulation_corrected_delays.get(comp_name, 0.0))

    Delay_PreAssembly = max(corrected_delays_to_consider)

    # Calculate Delay_Corrected_Assembly based on Delay_PreAssembly + Delay_Assembly
    Delay_Assembly = current_simulation_delays.get('Assembly', 0.0)
    Buffer_Assembly = globals().get('Buffer_Assembly', 0.0)

    combined_original_delay_for_assembly = Delay_PreAssembly + Delay_Assembly

    if combined_original_delay_for_assembly <= Buffer_Assembly:
        Delay_Corrected_Assembly = combined_original_delay_for_assembly
    else:
        if Buffer_Assembly <= 0:
            Delay_Corrected_Assembly = combined_original_delay_for_assembly
        else:
            random_beta_val_assembly = np.random.beta(a=1, b=3)
            value_to_subtract_assembly = Buffer_Assembly + random_beta_val_assembly * Buffer_Assembly
            corrected_delay_assembly = max(0.0, combined_original_delay_for_assembly - value_to_subtract_assembly)
            Delay_Corrected_Assembly = round(corrected_delay_assembly, 1)
            all_value_to_subtract_assembly.append(value_to_subtract_assembly) # Collect for plotting

    current_simulation_corrected_delays['Assembly'] = Delay_Corrected_Assembly

    # Define Delay_Transportation and Delay_Units
    distance_threshold_km = 500
    available_transport_modes = []
    if distance_km < distance_threshold_km:
      available_transport_modes = ['truck', 'train']
    else:
      available_transport_modes = ['truck', 'train', 'ship']

    selected_transport_method = None
    # Define delay thresholds to influence transport choice
    low_delay_threshold = 5 # Example: if delay <= 5 days, considered low
    high_delay_threshold = 10 # Example: if delay > 10 days, considered high

    if Delay_Corrected_Assembly <= low_delay_threshold:
      if 'ship' in available_transport_modes:
        selected_transport_method = 'ship'
      elif 'train' in available_transport_modes:
        selected_transport_method = 'train'
      else:
        selected_transport_method = 'truck' # Fallback
    elif Delay_Corrected_Assembly > high_delay_threshold:
      if 'truck' in available_transport_modes:
          selected_transport_method = 'truck'
      elif 'train' in available_transport_modes:
          selected_transport_method = 'train'
      elif 'ship' in available_transport_modes:
          selected_transport_method = 'ship'
    else: # Medium delay
      if 'train' in available_transport_modes:
          selected_transport_method = 'train'
      elif 'truck' in available_transport_modes:
          selected_transport_method = 'truck'
      elif 'ship' in available_transport_modes:
          selected_transport_method = 'ship'

    transport_file = ''
    if selected_transport_method == 'truck':
        transport_file = 'CAMION - Ritardi.xlsx'
    elif selected_transport_method == 'train':
        transport_file = 'TRENO - Ritardi.xlsx'
    elif selected_transport_method == 'ship':
        transport_file = 'NAVE - Ritardi.xlsx'

    Delay_Transportation = 0.0 # Default to 0 (no delay)
    try:
        transport_df = pd.read_excel(transport_file)
        # Find the row with 'Distanza (km)' closest to distance_km
        transport_df['distance_diff'] = abs(transport_df['Distanza (km)'] - distance_km)
        closest_row = transport_df.loc[transport_df['distance_diff'].idxmin()]

        # Get the probability of error
        probability_of_error_raw = closest_row['Percentuale']
        if isinstance(probability_of_error_raw, str):
            try:
                # Try with '%' first
                probability_of_error = float(probability_of_error_raw.strip('%').replace(',', '.')) / 100
            except ValueError:
                # If that fails, try as a direct number that needs division by 100
                probability_of_error = float(probability_of_error_raw.replace(',', '.')) / 100
        else:
            # Assume it's already a number (int or float) and convert to probability
            probability_of_error = float(probability_of_error_raw) / 100

        # Simulate the error using a beta function if an error occurs
        if np.random.rand() <= probability_of_error:
            min_val_h = closest_row['Minimo (h)']
            mean_h = closest_row['Medio (h)']
            max_val_h = closest_row['Massimo (h)']

            # Convert hours to days
            min_val_days = min_val_h / 24
            mean_days = mean_h / 24
            max_val_days = max_val_h / 24

            if min_val_days == max_val_days:
                Delay_Transportation = round(float(min_val_days), 1)
            else:
                kappa_transport = 13 # Using same kappa as generate_delay function
                normalized_mean_transport = (mean_days - min_val_days) / (max_val_days - min_val_days)
                epsilon_transport = 1e-6
                alpha_transport = normalized_mean_transport * kappa_transport + epsilon_transport
                beta_param_transport = (1 - normalized_mean_transport) * kappa_transport + epsilon_transport

                beta_val_transport = np.random.beta(alpha_transport, beta_param_transport)
                generated_val_transport = min_val_days + beta_val_transport * (max_val_days - min_val_days)
                Delay_Transportation = round(np.clip(generated_val_transport, min_val_days, max_val_days), 1)
        else:
            Delay_Transportation = 0.0

    except FileNotFoundError:
        print(f"Warning: {transport_file} not found. Setting Delay_Transportation to 0 for this simulation.")
        Delay_Transportation = 0.0
    except Exception as e:
        print(f"Error processing {transport_file}: {e}. Setting Delay_Transportation to 0 for this simulation.")
        Delay_Transportation = 0.0
    # --- END NEW LOGIC ---

    Delay_Units = Delay_Corrected_Assembly + Delay_Transportation

    # Define Delay_OnSite
    Delay_Brackets = current_simulation_corrected_delays.get('Brackets', 0.0)
    Delay_ChemicalAnchors = current_simulation_corrected_delays.get('ChemicalAnchors', 0.0)

    Delay_OnSite = max(Delay_Units, Delay_Brackets, Delay_ChemicalAnchors)

    # New logic for Delay_BeforeInstallation
    if Delay_OnSite == Delay_Units:
        Delay_BeforeInstallation = max(0.0, Delay_OnSite - recovery_installation)
    else:
        Delay_BeforeInstallation = Delay_OnSite

    # Retrieve Delay_Installation from simulation results for 'Installation' component
    # It's now calculated and its corrected value is its original value (no buffer applied).
    Delay_Installation = current_simulation_corrected_delays.get('Installation', 0.0)

    # Define Delay_Total
    Delay_Total = Delay_BeforeInstallation + Delay_Installation

    # Store all results for the current simulation
    simulation_result = {
        'simulation_number': s_idx + 1,
        'component_delays': current_simulation_delays,
        'component_corrected_delays': current_simulation_corrected_delays,
        'Delay_PreAssembly': Delay_PreAssembly,
        'Selected_Transport_Method': selected_transport_method, # Added for printing
        'Delay_Transportation': Delay_Transportation,
        'Delay_Units': Delay_Units,
        'Delay_OnSite': Delay_OnSite,
        'Delay_BeforeInstallation': Delay_BeforeInstallation,
        'Delay_Installation': Delay_Installation,
        'Delay_Total': Delay_Total
    }
    all_simulations_results.append(simulation_result)

print("\nAll simulations completed. Printing results for each simulation:")

# Print only the first two and last two simulation results if num_simulations > 4
results_to_print = []
if num_simulations <= 4:
    results_to_print = all_simulations_results
else:
    results_to_print.extend(all_simulations_results[:2]) # First two
    results_to_print.append({'message': f'... {num_simulations - 4} simulations omitted ...'})
    results_to_print.extend(all_simulations_results[-2:]) # Last two

for result in results_to_print:
    if 'message' in result:
        print(result['message'])
    else:
        print(f"\n--- Results for Simulation {result['simulation_number']} ---")
        print("Individual component delays:")
        for component_name in component_names_list:
            original_delay_value = result['component_delays'].get(component_name, 'Not calculated')
            corrected_delay_value = result['component_corrected_delays'].get(component_name, 'Not calculated')
            if component_name == 'Installation':
                # User request: "che non venga neanche stampato" for corrected delay
                print(f'  Delay_{component_name}: {original_delay_value}')
            else:
                print(f'  Delay_{component_name}: {original_delay_value}, Delay_Corrected_{component_name}: {corrected_delay_value}')
        print(f"Calculated Delay_PreAssembly: {result['Delay_PreAssembly']}")
        print("## Resources") # Added this line
        print(f"Selected Transport Method: {result['Selected_Transport_Method']}") # Printing here as requested
        print(f"Calculated Delay_Transportation: {result['Delay_Transportation']}")
        print(f"Calculated Delay_Units: {result['Delay_Units']}")
        print(f"Calculated Delay_OnSite: {result['Delay_OnSite']}")
        print(f"Calculated Delay_BeforeInstallation: {result['Delay_BeforeInstallation']}")
        print(f"Calculated Delay_Installation: {result['Delay_Installation']}")
        print(f"Calculated Delay_Total: {result['Delay_Total']}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- Interactive inputs for visualization ---
target_delay_input = 0 # @param {type:"number"} # Set 0 for no input
target_percentage_input = 80 # @param {type:"number"} # Set 0 for no input (e.g., 90 for 90%)

print("\n--- Visualizing Interactive Targets on CDF ---")

# Extract all Delay_Total values from the results of the simulation cell
delay_totals = [result['Delay_Total'] for result in all_simulations_results]
sorted_delays = np.sort(delay_totals)
n = len(sorted_delays)

x_vals = np.concatenate([[0], sorted_delays])
y_vals = np.concatenate([[0], np.arange(1, n + 1) / n])

# --- Probability of zero delay (for the highlighted band) ---
prob_zero = float(np.sum(np.array(sorted_delays) == 0)) / n

dark_blue = '#4887DC'
red_color = '#102C53'

fig, ax = plt.subplots(figsize=(10, 6))

# --- Draw CDF as step function anchored at (0, 0) ---
ax.step(x_vals, y_vals, where='post', color=dark_blue, linewidth=2)

# --- Light fill and dashed line highlighting P(Delay = 0) ---
ax.axhspan(0, prob_zero, alpha=0.10, color='gray', zorder=0,
           label=f'P($Delay_{{total}}$ = 0) = {prob_zero:.1%}')
ax.axhline(y=prob_zero, color='gray', linestyle='--', linewidth=0.5, alpha=0.85)

# --- Interactive target lines: both dashed, both reaching their respective axis ---
if target_delay_input > 0:
    cumulative_prob = float(np.sum(sorted_delays <= target_delay_input)) / n
    # Vertical dashed line from x-axis (y=0) up to intersection point
    ax.plot([target_delay_input, target_delay_input], [0, cumulative_prob],
            color=red_color, linestyle='--', linewidth=1.5)
    # Horizontal dashed line from y-axis (x=0) to intersection point
    ax.plot([0, target_delay_input], [cumulative_prob, cumulative_prob],
            color=red_color, linestyle='--', linewidth=1.5)
    ax.scatter(target_delay_input, cumulative_prob, color=red_color, zorder=5, s=50)


    x_ticks = sorted(set(list(ax.get_xticks()) + [target_delay_input]))
    ax.set_xticks(x_ticks)
    x_labels = [f'{t:.1f}' if abs(t - target_delay_input) < 1e-9 else str(int(t)) for t in x_ticks]
    ax.set_xticklabels(x_labels)
    for label, tick_val in zip(ax.get_xticklabels(), x_ticks):
        if abs(tick_val - target_delay_input) < 1e-9:
            label.set_color(red_color)
            label.set_fontweight('bold')


    y_ticks = sorted(set(list(ax.get_yticks()) + [cumulative_prob]))
    ax.set_yticks(y_ticks)
    y_labels = [f'{t:.0%}' for t in y_ticks]
    ax.set_yticklabels(y_labels)
    for label, tick_val in zip(ax.get_yticklabels(), y_ticks):
        if abs(tick_val - cumulative_prob) < 1e-9:
            label.set_color(red_color)
            label.set_fontweight('bold')

elif target_percentage_input > 0 and target_percentage_input <= 100:
    target_percentile_value = float(np.percentile(sorted_delays, target_percentage_input))
    prob = target_percentage_input / 100.0
    # Vertical dashed line from x-axis (y=0) up to intersection point
    ax.plot([target_percentile_value, target_percentile_value], [0, prob],
            color=red_color, linestyle='--', linewidth=1.5)
    # Horizontal dashed line from y-axis (x=0) to intersection point
    ax.plot([0, target_percentile_value], [prob, prob],
            color=red_color, linestyle='--', linewidth=1.5)
    ax.scatter(target_percentile_value, prob, color=red_color, zorder=5, s=50)


    x_ticks = sorted(set(list(ax.get_xticks()) + [target_percentile_value]))
    ax.set_xticks(x_ticks)
    x_labels = [f'{t:.1f}' if abs(t - target_percentile_value) < 1e-9 else str(int(t)) for t in x_ticks]
    ax.set_xticklabels(x_labels)
    for label, tick_val in zip(ax.get_xticklabels(), x_ticks):
        if abs(tick_val - target_percentile_value) < 1e-9:
            label.set_color(red_color)
            label.set_fontweight('bold')


    y_ticks = sorted(set(list(ax.get_yticks()) + [prob]))
    ax.set_yticks(y_ticks)
    y_labels = [f'{t:.0%}' for t in y_ticks]
    ax.set_yticklabels(y_labels)
    for label, tick_val in zip(ax.get_yticklabels(), y_ticks):
        if abs(tick_val - prob) < 1e-9:
            label.set_color(red_color)
            label.set_fontweight('bold')


ax.set_title(r"Cumulative Distribution Function (CDF) of $Delay_{total}$")
ax.set_xlabel('$Delay_{total}$ (days)')
ax.set_ylabel('Supply Chain Reliability Index (SCR)')
ax.set_xlim(left=0)
ax.set_ylim(bottom=0, top=1.05)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```
